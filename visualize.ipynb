```python
import os
import torch
os.environ['TORCH'] = torch.__version__
print(torch.__version__)

%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
from torch_geometric.utils import to_networkx
```

```python
def visualize_graph(G, color):
    plt.figure(figsize=(10,10))
    plt.xticks([])
    plt.yticks([])
    nx.draw_networkx(G, pos=nx.spring_layout(G, seed=42), with_labels=False,
                     node_color=color, cmap="Set2", node_size=10)
    plt.savefig('Images/graph.png')
    plt.show()
```

```python
usu_graph = torch.load('Graph/usu_graph.pth')
```

```python
G = to_networkx(usu_graph, to_undirected=True)
visualize_graph(G, color=usu_graph.y)
```

```python
print('Graph Eigenschaften')
print('==============================================================')

print(f'Anzahl der Knoten: {usu_graph.num_nodes}')
print(f'Anzahl der Kanten: {usu_graph.num_edges}')
print(f'Durchschnittlicher Knotengrad: {usu_graph.num_edges / usu_graph.num_nodes:.2f}')
print(f'Enthlt isolierte Knoten: {usu_graph.has_isolated_nodes()}')
print(f'Enthlt Schleifen: {usu_graph.has_self_loops()}')
print(f'Ungerichtet: {usu_graph.is_undirected()}')
```
